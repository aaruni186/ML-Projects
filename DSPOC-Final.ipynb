```python
# Reading the data sets
import pandas as pd
training_data = pd.read_excel("Training-Data-Sets.xlsx")
test_data  = pd.read_excel("Test dataset v1.xlsx")
```

```python
# Get summary statistics for the taining data set
training_data.describe().T
```

```python
# Get summary statistics for the test data set
test_data.describe().T
```

```python
# Setting the Day column as the index
training_data.set_index("Day", inplace=True)
```

```python
# Getting the data types of all the columns
training_data.dtypes
```

```python
# All the columns are of a numeric type. Separating out the dependant variable
EQ = training_data["EQ"]

# Checking the data distribution of the dependant variable
EQ.plot.kde();
```

```python
# Extracting the features.
regressors = training_data.drop("EQ",axis=1)
regressors.info()
```

```python
# Applying Principal Component Analysis to reduce the number of features
from sklearn.decomposition import PCA

pca = PCA(n_components = 10)
filtered_features = pca.fit_transform(regressors)
filtered_features.shape
```

```python
# Scaling the features and dependant variables
from  sklearn.preprocessing import StandardScaler
feature_scaler = StandardScaler()
scaled_features  = feature_scaler.fit_transform(filtered_features)

EQ_scaler = StandardScaler()
scaled_EQ = EQ_scaler.fit_transform(EQ.values.reshape(-1,1))

(scaled_features, scaled_EQ)
```

```python
# Scaling the test values of the target variable
test_scaled_EQ = EQ_scaler.transform(test_data["EQ"].values.reshape(-1,1))
test_scaled_EQ.shape
```

```python
# Applying PCA transformation on the test data set
test_features = test_data.drop("EQ",axis=1).set_index("Period")
test_features = pca.transform(test_features)
test_features.shape
```

```python
# Scaling the test features
test_scaled_features = feature_scaler.transform(test_features)
test_scaled_features.shape
```

```python
# Creating the data set for the Bayesian Regression input
import numpy as np
data = np.concatenate((scaled_EQ,scaled_features), axis =1)
data = pd.DataFrame(data)
col_names = ["EQ"]

for id in range(1,scaled_features.shape[1]+ 1):
    col_names.append("C" + str(id))
    
data.columns = col_names
data
```

```python
# Creating the Bayesian Regression formula using the column names of the features
formula = 'EQ ~ ' + ' + '.join(['%s' % variable for variable in data.columns[1:]])
formula
```

```python
# Creating the Bayesian Regression model
import pymc3 as pm


with pm.Model() as normal_model:
    
    # The prior for the model parameters will be a normal distribution
    family = pm.glm.families.Normal()
    
    # Creating the model requires a formula and data (and optionally a family)
    pm.GLM.from_formula(formula, data = data , family = family)
    
    # Perform Markov Chain Monte Carlo sampling
    normal_trace = pm.sample(draws=2000, chains = 2, tune = 500)
```

```python
# Ploting the Bayesian Regression model parameter distributions
pm.plot_posterior(normal_trace, figsize = (14, 14));
```

```python
# Displaying the Bayesian Regression parameter summary
br_results = pm.summary(normal_trace)
br_results
```

```python
# Extracting the mean intercept value
intercept = br_results.iloc[0,0]
intercept_vector = np.repeat(np.array(intercept),test_scaled_EQ.size) 
intercept_vector
```

```python
# Extracting the mean weights
weights = br_results.iloc[1:-1,0].values
weights
```

```python
# Calculating the predicted target variables
br_pred_EQ = np.dot(test_scaled_features,weights) + intercept_vector
br_pred_EQ
```

```python
# Calculating the MAPE value for Bayesian Regression
mape = np.absolute((test_scaled_EQ.reshape(1,-1)-br_pred_EQ)/test_scaled_EQ.reshape(1,-1)).mean()
mape
```

```python
# Creating a Random Forest Regression model and predicting the target variable values
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators = 100, random_state=42)
rf_model.fit(scaled_features,scaled_EQ)

rf_pred_EQ = rf_model.predict(test_scaled_features)

rf_pred_EQ
```

```python
# Calculating the MAPE for Random Forest Regression
rf_mape = np.absolute((test_scaled_EQ-rf_pred_EQ)/test_scaled_EQ).mean()
rf_mape
```
